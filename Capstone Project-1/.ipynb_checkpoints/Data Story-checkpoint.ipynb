This capstone project is examining the College Scorecard data to predict the factors that contribute to student landing in a high income job. 
 - How does SAT score impact the mean earnings?
 - How does the tuition fee impact the mean earnings?
 - Is there a relation between faculty salary and the course completion?
 - Does a particular gender or race have more completion rates/earnings than the other?
 - Does the college impact the mean earnings?

```python
scorecard = pd.read_excel('data\Scorecard_2014_15.xlsx')
```

```python
scorecard.head()
```

```python
scorecard.plot(kind='scatter',x='COSTT4_A',y='MD_EARN_WNE_P6')
plt.xlabel('Average Cost of Attendance')
plt.ylabel('6yr Mean Earnings')
plt.plot()
plt.show()
```

```python
scorecard.plot(kind='scatter',x='NPT4',y='MD_EARN_WNE_P6')
plt.xlabel('Net Price')
plt.ylabel('6yr Mean Earnings')
plt.plot()
```

```python
scorecard.plot(kind='scatter',x='AVGFACSAL',y='MD_EARN_WNE_P6')
plt.xlabel('Faculty Salary')
plt.ylabel('6yr Mean Earnings')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.plot()
plt.show()
```

```python
scorecard.plot(kind='scatter',x='AVGFACSAL',y='MD_EARN_WNE_P6')
plt.xlabel('Faculty Salary')
plt.ylabel('6yr Mean Earnings')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.plot()
plt.show()
```

```python
scorecard['MD_EARN_WNE_P6'].hist(bins=20)
plt.xlabel("Mean Earnings 6yr")
plt.ylabel('Frequency')
plt.title('Histogram of Mean Earnings')
plt.show()
```
