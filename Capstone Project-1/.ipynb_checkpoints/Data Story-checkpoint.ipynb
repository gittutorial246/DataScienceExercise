```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from IPython.core.display import HTML
css = open('data\style-table.css').read() 
HTML('<style>{}</style>'.format(css))
```

This capstone project is examining the College Scorecard data to predict the factors that contribute to student landing in a high income job. 
 - How does SAT score impact the mean earnings?
 - How does the tuition fee impact the mean earnings?
 - Is there a relation between faculty salary and the course completion?
 - Does a particular gender or race have more completion rates/earnings than the other?
 - Does the college impact the mean earnings?

```python
scorecard = pd.read_excel('data\Scorecard_2014_15.xlsx')
```

```python
scorecard.head()
```

## COST
You can see that none of the cost metrics are correlated with mean earnings.There is moderate correlation between the out of state tuition fees and mean earnings. 
TODO : See if you can get data on out of state tuition fees from other cohorts

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
fig.set_figheight(10)
fig.set_figwidth(15)

scorecard.plot(kind='scatter',x='MN_EARN_P10',y='AVGCOST',ax=axes[0,0])
axfig = axes[0,0]
axfig.set_title('Avg cost')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('Avg cost of attendance')
axfig.set_xlim(left=20000,right=80000)
axfig.set_ylim(bottom=1000,top=70000)


scorecard.plot(kind='scatter',x='MN_EARN_P10',y='NPT4',ax=axes[0,1])
axfig = axes[0,1]
axfig.set_title('Avg net price')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('Avg net price')
axfig.set_xlim(left=10000,right=80000)
axfig.set_ylim(bottom=0,top=40000)

scorecard.plot(kind='scatter',x='MN_EARN_P10',y='TUITIONFEE_OUT',ax=axes[0,2])
axfig = axes[0,2]
axfig.set_title('Tuition fee OUT')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('Tuition fee OUT')
axfig.set_xlim(left=10000,right=80000)
axfig.set_ylim(bottom=0,top=60000)

scorecard.plot(kind='scatter',x='MN_EARN_P10',y='TUITIONFEE_IN',ax=axes[1,0])
axfig = axes[1,0]
axfig.set_title('Tuition fee')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('Tuition fee')
axfig.set_xlim(left=10000,right=80000)
axfig.set_ylim(bottom=0,top=60000)

scorecard.plot(kind='scatter',x='NUM_EARN_P10',y='NUM4',ax=axes[1,1])
axfig = axes[1,1]
axfig.set_title('No. of students')
axfig.set_xlabel('10 yr count of earnings')
axfig.set_ylabel('Title IV Students')
axfig.set_xlim(left=0,right=7500)
axfig.set_ylim(bottom=-500,top=4000)

plt.tight_layout()
plt.show()
```

## ADMISSIONS
You can see that SAT and ACT scores are correlated to mean earnings after 10 years. The admission rate is not correlated.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)
scorecard.plot(kind='scatter',x='MN_EARN_P10',y='ACTCMMID',ax=axes[0,0])
axfig = axes[0,0]
axfig.set_title('ACT')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('ACT')
axfig.set_xlim(left=20000,right=80000)
axfig.set_ylim(bottom=10,top=40)


scorecard.plot(kind='scatter',x='MN_EARN_P10',y='ADM_RATE',ax=axes[0,1])
axfig = axes[0,1]
axfig.set_title('Overall admission Rate')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('Overall admission Rate')
axfig.set_xlim(left=10000,right=80000)
axfig.set_ylim(bottom=0,top=1)


scorecard.plot(kind='scatter',x='MN_EARN_P10',y='SAT_AVG',ax=axes[1,0])
axfig = axes[1,0]
axfig.set_title('SAT')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('SAT')
axfig.set_xlim(left=25000,right=80000)
axfig.set_ylim(bottom=600,top=1500)

scorecard.plot(kind='scatter',x='MN_EARN_P10',y='SAT_AVG_ALL',ax=axes[1,1])
axfig = axes[1,1]
axfig.set_title('SAT')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('SAT')
axfig.set_xlim(left=25000,right=80000)
axfig.set_ylim(bottom=600,top=1500)

plt.tight_layout()
plt.show()
```

## SCHOOL
You can see that Faculty salary is reasonably correlated with Mean Earnings. Institutional expense is moderately correlated.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)
scorecard.plot(kind='scatter',x='MN_EARN_P10',y='AVGFACSAL',ax=axes[0,0])
axfig = axes[0,0]
axfig.set_title('Faculty Salary')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('Faculty Salary')
axfig.set_xlim(left=20000,right=80000)
axfig.set_ylim(bottom=0,top=15000)


scorecard.plot(kind='scatter',x='MN_EARN_P10',y='INEXPFTE',ax=axes[0,1])
axfig = axes[0,1]
axfig.set_title('Institutional Expense per student')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('Expense')
axfig.set_xlim(left=10000,right=80000)
axfig.set_ylim(bottom=0,top=15000)


scorecard.plot(kind='scatter',x='MN_EARN_P10',y='TUITFTE',ax=axes[1,0])
axfig = axes[1,0]
axfig.set_title('Net tuition revenue per student')
axfig.set_xlabel('10 yr Mean earnings')
axfig.set_ylabel('Tuition Revenue')
axfig.set_xlim(left=20000,right=80000)
axfig.set_ylim(bottom=0,top=30000)

plt.tight_layout()
plt.show()
```
