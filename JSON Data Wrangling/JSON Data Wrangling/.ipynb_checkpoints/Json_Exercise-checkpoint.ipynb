## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
json_normalize(data, 'counties', ['state', 'shortname'])
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Find the 10 countries with most projects
import pandas as pd
world_bank = pd.read_json('data/world_bank_projects.json')

project_count = world_bank.groupby(['countrycode','countryname'])['project_name'].agg(pd.Series.nunique).sort_values(ascending=False).head(10)
project_count.columns = ['countrycode','countryname','count']
project_count
```

```python
# Find the top 10 major project themes (using column 'mjtheme_namecode')
json_str = json.load((open('data/world_bank_projects.json')))
project_themes = json_normalize(json_str,'mjtheme_namecode',['project_name'])

project_themes = project_themes.drop_duplicates() # record count changed from 1499 to 1085
# Top 10 themes based on occurences across projects
project_themes.groupby(['name','code']).size().sort_values(ascending=False).head(10)
```

```python
project_themes.sort_values('code')
```

```python
#Create a dataframe with the missing names filled in.
#create a lookup with just code and name
theme_lookup = project_themes[~(project_themes.name=='')][['code','name']].drop_duplicates()

themes_filled = pd.merge(project_themes, theme_lookup, on='code')

#drop the duplicate column
themes_filled = themes_filled.drop(['name_x'],axis=1)
themes_filled.columns = ['Theme_Code','Project_Name','Theme_Name']
# Themes will values filled, note the difference in counts
themes_filled['Theme_Name'].value_counts().head(10)
```
